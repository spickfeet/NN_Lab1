```python
import numpy as np

np.random.seed(42)

R = np.array([
    [5, 3, 0, 1, 4],
    [4, 0, 2, 0, 5],
    [0, 1, 5, 3, 0],
    [2, 0, 0, 4, 3],
    [3, 4, 1, 0, 2]
], dtype=float)

print("Исходная матрица оценок R (5×5):")
print(R)
print(f"Количество ненулевых элементов: {np.count_nonzero(R)} из {R.size}")
print()


n_users, n_items = R.shape 
d = 3

P = np.random.randn(n_users, d) * 0.1 
Q = np.random.randn(d, n_items) * 0.1

print(f"Матрица P (пользователи × латентные признаки): форма {P.shape}")
print(P)

print(f"Матрица Q (латентные признаки × товары): форма {Q.shape}")
print(Q)


R_pred = P @ Q

print("Предсказанная матрица R_pred = P × Q:")
print(np.round(R_pred, 4))
print()

mask = R != 0

R_observed = R[mask]
R_pred_observed = R_pred[mask]


mse = np.mean((R_observed - R_pred_observed) ** 2)

print("Результаты")
print(f"Количество известных оценок: {np.sum(mask)}")
print(f"Среднеквадратическая ошибка (MSE) по ненулевым элементам: {mse:.6f}")
print(f"RMSE: {np.sqrt(mse):.6f}")
print()

print("Детали по известным оценкам:")
print(f"{'Индекс':<10} {'R (истина)':<12} {'R_pred':<12} {'Ошибка^2':<12}")

indices = np.where(mask)
for i, (row, col) in enumerate(zip(indices[0], indices[1])):
    true_val = R[row, col]
    pred_val = R_pred[row, col]
    error_sq = (true_val - pred_val) ** 2
    print(f"[{row},{col}]:      {true_val:<12.4f} {pred_val:<12.4f} {error_sq:<12.6f}")
```
